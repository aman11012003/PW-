### 1. **What is hypothesis testing in statistics?**  
Hypothesis testing is a statistical method used to determine if there is enough evidence to reject a null hypothesis.  
It involves comparing observed data against an expected outcome under a certain assumption.  
A hypothesis test evaluates whether an effect or relationship exists in a population.  
It uses statistical tests like t-tests, z-tests, and chi-square tests.  
The conclusion is based on a significance level and p-value.

---

### 2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
The **null hypothesis (H₀)** states that there is no effect or difference in a population.  
The **alternative hypothesis (H₁)** suggests that there is a significant effect or difference.  
Rejecting H₀ means the data supports H₁, while failing to reject H₀ means there is insufficient evidence.  
H₀ is assumed to be true until proven otherwise by statistical evidence.  
Statistical tests are performed to determine if we reject or fail to reject H₀.


---

### 3. **What is the significance level in hypothesis testing, and why is it important?**  
The significance level (α) is the probability of rejecting a true null hypothesis (Type I error).  
Common values are **0.05 (5%)** or **0.01 (1%)**, indicating the confidence in the test result.  
A lower α reduces the chance of making a Type I error but increases the chance of Type II error.  
It determines the critical value for decision-making in hypothesis tests.  
A significance level of **0.05** means there's a **5% risk** of concluding an effect exists when it doesn’t.


---
### 4. **What does a P-value represent in hypothesis testing?**  
The P-value indicates the probability of observing the data if the null hypothesis is true.  
A **low P-value (≤ α)** suggests strong evidence against H₀, leading to its rejection.  
A **high P-value (> α)** indicates weak evidence, meaning we fail to reject H₀.  
It quantifies the level of statistical significance of the test result.  
P-values are compared with the significance level to make conclusions.

---

### 5. **How do you interpret the P-value in hypothesis testing?**  
If **P ≤ α**, reject the null hypothesis (strong evidence against H₀).  
If **P > α**, fail to reject H₀ (insufficient evidence to support H₁).  
For example, if **P = 0.03** and **α = 0.05**, we reject H₀.  
A smaller P-value means stronger evidence against the null hypothesis.  
A very large P-value suggests the observed data is consistent with H₀.

--

### 6. **What are Type 1 and Type 2 errors in hypothesis testing?**  
A **Type I error** occurs when H₀ is rejected even though it is true (false positive).  
A **Type II error** happens when H₀ is not rejected even though it is false (false negative).  
Type I errors are controlled by the significance level (α), usually 5%.  
Reducing Type I errors increases Type II errors and vice versa.  
Type I is more serious in critical scenarios like medical testing.


---

### 7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**  
A **one-tailed test** checks for an effect in only one direction (e.g., greater than or less than).  
A **two-tailed test** checks for an effect in both directions (e.g., different but not specified how).  
One-tailed tests have more statistical power but require strong justification.  
Two-tailed tests are used when the effect could go in either direction.  
Example: Testing if a new drug works **better (one-tailed)** or **differently (two-tailed)** than an existing one.

---

### 8. **What is the Z-test, and when is it used in hypothesis testing?**  
A Z-test is used to compare population means when the sample size is large (n > 30).  
It assumes that the data follows a normal distribution.  
It is used when population variance is known.  
Common applications include testing proportions and comparing means.  
Example: Testing whether the average height of a group differs from the population mean.


---
### 9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**  
The **Z-score** is calculated as:  
\[
Z = \frac{X - \mu}{\sigma}
\]  
where **X** is the sample mean, **μ** is the population mean, and **σ** is the standard deviation.  
It represents how many standard deviations a data point is from the mean.  
A **higher Z-score** indicates the data point is far from the mean.  
Z-scores are used to determine probabilities in a normal distribution.


----
### 10. **What is the T-distribution, and when should it be used instead of the normal distribution?**  
The **T-distribution** is used when the sample size is small (n < 30).  
It accounts for extra uncertainty due to unknown population variance.  
It is similar to the normal distribution but has heavier tails.  
It is used in **t-tests** for hypothesis testing of small sample means.  
The larger the sample size, the more the t-distribution resembles the normal distribution.


----
### 11. **What is the difference between a Z-test and a T-test?**  
- **Z-test** is used for **large samples (n > 30)** and when **population variance is known**.  
- **T-test** is used for **small samples (n < 30)** and when **population variance is unknown**.  
- Z-test follows a **normal distribution**, whereas T-test follows a **T-distribution**.  
- T-test has **heavier tails** (more variability) compared to Z-test.  
- As sample size increases, T-test results approach Z-test results.

---

### 12. **What is the T-test, and how is it used in hypothesis testing?**  
- A T-test is used to **compare the means of two groups** when sample size is small.  
- It determines whether the observed difference is **statistically significant**.  
- Assumes data is **normally distributed** and samples are **independent**.  
- Used in **single-sample, independent, and paired-sample** scenarios.  
- Example: Comparing **test scores of students from two different schools**.

---

### 13. **What is the relationship between Z-test and T-test in hypothesis testing?**  
- Both tests help determine if a sample **differs significantly** from a population.  
- Z-test is used for **large samples with known variance**, while T-test is used for **small samples with unknown variance**.  
- T-test is more **conservative** due to its heavier tails.  
- As sample size increases, **T-distribution approximates the normal distribution**, making T-test similar to Z-test.  
- Both tests use **critical values and p-values** to make decisions.

---

### 14. **What is a confidence interval, and how is it used to interpret statistical results?**  
- A **confidence interval (CI)** is a range of values that likely contains the **true population parameter**.  
- It is calculated as:  
  \[
  CI = \bar{X} \pm (Z \text{ or } t) \times \frac{\sigma}{\sqrt{n}}
  \]
- **Wider CI** means more uncertainty; **narrower CI** means more precision.  
- **95% CI** means that if we repeat the experiment, the true value will be in this range **95% of the time**.  
- Example: "The average height of students is **160 cm ± 5 cm** (95% CI)".

---

### 15. **What is the margin of error, and how does it affect the confidence interval?**  
- The **margin of error (MoE)** is the amount added/subtracted to the mean to create a confidence interval.  
- It depends on the **sample size (n), standard deviation (σ), and confidence level**.  
- **Larger samples → smaller MoE** → more precise results.  
- **Higher confidence levels (e.g., 99%) → larger MoE** → more uncertainty.  
- Reducing MoE requires **larger sample sizes** or **lower confidence levels**.


---
### 16. **How is Bayes' Theorem used in statistics, and what is its significance?**  
- **Bayes' Theorem** calculates the **probability of an event based on prior knowledge**.  
- Formula:  
  \[
  P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
  \]
- Used in **spam filtering, medical diagnosis, and machine learning**.  
- Helps **update beliefs** as new evidence is available.  
- Example: If a medical test is **90% accurate**, Bayes' Theorem adjusts the probability of disease **based on test results**.


---
### 17. **What is the Chi-square distribution, and when is it used?**  
- The **Chi-square distribution** is used to test relationships between categorical variables.  
- It is **positively skewed** and used in **goodness-of-fit and independence tests**.  
- Requires **non-negative values** and assumes **independent observations**.  
- Used to **compare observed vs. expected frequencies**.  
- Example: Checking if **gender affects voting preference**.

----

### 18. **What is the Chi-square goodness of fit test, and how is it applied?**  
- The **goodness of fit test** checks if **observed data matches an expected distribution**.  
- Formula:  
  \[
  \chi^2 = \sum \frac{(O - E)^2}{E}
  \]
- If **calculated χ² > critical value**, reject H₀ (data does not fit).  
- Used in **market research, genetics, and quality control**.  
- Example: Checking if **dice are fair or biased**.


----

### 19. **What is the F-distribution, and when is it used in hypothesis testing?**  
- The **F-distribution** is used to compare **variances of two populations**.  
- It is **positively skewed** and depends on **degrees of freedom**.  
- Used in **ANOVA (Analysis of Variance) and F-tests**.  
- Requires **independent, normally distributed samples**.  
- Example: Testing if **three teaching methods result in different student performance**


---
### 20. **What is an ANOVA test, and what are its assumptions?**  
- **ANOVA (Analysis of Variance)** tests if **three or more group means are different**.  
- Assumptions:  
  1. **Normality**: Data follows a normal distribution.  
  2. **Independence**: Samples are independent of each other.  
  3. **Homogeneity of variance**: Variance is similar across groups.  
  4. **Random sampling**: Data is randomly collected.  
  5. **No extreme outliers**: Outliers can distort results.


  ---

### 21. **What are the different types of ANOVA tests?**  
- **One-way ANOVA**: Tests differences among **one independent variable** (e.g., test scores across schools).  
- **Two-way ANOVA**: Examines effects of **two independent variables** (e.g., test scores based on gender and school).  
- **Repeated Measures ANOVA**: Compares means when the same subjects are measured multiple times.  
- **MANOVA (Multivariate ANOVA)**: Used when there are **multiple dependent variables**.  
- **ANCOVA (Analysis of Covariance)**: ANOVA with a **covariate** to control for an extra variable.

----

### 22. **What is the F-test, and how does it relate to hypothesis testing?**  
- An **F-test** compares **two population variances** or **multiple group means**.  
- Formula:  
  \[
  F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}}
  \]
- Used in **ANOVA and regression analysis**.  
- If F **is significantly large**, we reject H₀ (variances are different).  
- Example: Comparing **two teaching methods' variability in student scores**.

```python
# Perform a Z-test for comparing a sample mean to a known population mean
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis (H0)")
    else:
        print("Fail to reject the null hypothesis (H0)")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(50, 5, 30)  # Sample data
z_test(sample_data, pop_mean=52, pop_std=5)
```

```python
# Perform a two-tailed Z-test and visualize the decision region

import matplotlib.pyplot as plt

def plot_z_test(z_stat):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.axvline(z_stat, color='r', linestyle='dashed', label=f'Z-score: {z_stat:.2f}')
    plt.axvline(-1.96, color='g', linestyle='dashed', label='Critical Value (-1.96)')
    plt.axvline(1.96, color='g', linestyle='dashed', label='Critical Value (1.96)')

    plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3)
    plt.title("Two-tailed Z-test Decision Region")
    plt.legend()
    plt.show()

z_stat, _ = ztest(sample, value=52)
plot_z_test(z_stat)
```

```python
# Visualizing Type 1 and Type 2 Errors

def plot_type_errors():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Null Distribution (H0)")

    plt.axvline(1.645, color='r', linestyle='dashed', label='Type 1 Error Threshold (α)')
    plt.axvline(-1.645, color='r', linestyle='dashed')

    plt.fill_between(x, y, where=(x > 1.645), color='red', alpha=0.3, label='Type 1 Error')
    plt.fill_between(x, y, where=(x < -1.645), color='red', alpha=0.3)

    plt.title("Type 1 and Type 2 Errors")
    plt.legend()
    plt.show()

plot_type_errors()
```

```python
# Function to Calculate Confidence Interval
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    margin = stats.t.ppf((1 + confidence) / 2, df=len(sample)-1) * std_err

    lower_bound = mean - margin
    upper_bound = mean + margin

    print(f"{confidence*100}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    return lower_bound, upper_bound

sample_data = np.random.normal(50, 5, 30)
confidence_interval(sample_data)
```

```python
#Calculate Margin of Error for a Given Confidence Level
import numpy as np
from scipy.stats import norm

def margin_of_error(sample, confidence=0.95):
    sample_mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    critical_value = norm.ppf((1 + confidence) / 2)
    margin = critical_value * std_err
    print(f"Margin of Error: {margin:.4f}")
    return margin

sample_data = np.random.normal(50, 5, 30)
margin_of_error(sample_data)
```

```python
# Implement Bayesian Inference using Bayes' Theorem
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    posterior_A = (likelihood_B_given_A * prior_A) / ((likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A))
    print(f"Posterior Probability: {posterior_A:.4f}")
    return posterior_A

bayes_theorem(prior_A=0.01, likelihood_B_given_A=0.9, prior_not_A=0.99, likelihood_B_given_not_A=0.05)
```

```python
#Perform a Chi-square test for independence between two categorical variables in PythonD


import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Creating a contingency table (Example: Survey of product preferences by gender)
data = np.array([[50, 30],  # Males
                 [20, 40]])  # Females

# Convert to a Pandas DataFrame
df = pd.DataFrame(data, columns=["Product A", "Product B"], index=["Male", "Female"])
print("Contingency Table:")
print(df)

# Performing the Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Display results
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=df.columns, index=df.index))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis. The two variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. The two variables are independent.")
```

```python
#Implement an F-test for Variance Comparison
from scipy.stats import f

def f_test(sample1, sample2):
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    f_stat = var1 / var2
    df1, df2 = len(sample1) - 1, len(sample2) - 1
    p_value = 1 - f.cdf(f_stat, df1, df2)

    print(f"F-statistic: {f_stat:.4f}, P-value: {p_value:.4f}")

sample1 = np.random.normal(50, 5, 30)
sample2 = np.random.normal(50, 10, 30)
f_test(sample1, sample2)
```

```python
#Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f

# Generate two random datasets with different variances
np.random.seed(42)
data1 = np.random.normal(50, 5, 30)  # Mean=50, Std=5
data2 = np.random.normal(50, 10, 30)  # Mean=50, Std=10

# Compute variances
var1 = np.var(data1, ddof=1)  # Sample variance
var2 = np.var(data2, ddof=1)

# Compute F-statistic
F_stat = var1 / var2
df1, df2 = len(data1) - 1, len(data2) - 1  # Degrees of freedom

# Compute P-value
p_value = 1 - f.cdf(F_stat, df1, df2)

# Print results
print(f"Variance of Dataset 1: {var1:.4f}")
print(f"Variance of Dataset 2: {var2:.4f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis. The variances are significantly different.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. The variances are similar.")

# Visualization: Boxplot for variance comparison
plt.figure(figsize=(8, 5))
sns.boxplot(data=[data1, data2])
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
plt.title("Variance Comparison using Boxplot")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
#Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulating observed data (e.g., number of customers choosing different products)
np.random.seed(42)
observed = np.random.randint(20, 100, size=5)  # Random observed frequencies for 5 categories

# Expected data assuming uniform distribution
expected = np.full(len(observed), np.mean(observed))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed, expected)

# Print results
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis. The observed distribution does not match the expected distribution.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. The observed distribution matches the expected distribution.")

# Visualization: Observed vs. Expected Frequencies
categories = [f"Category {i+1}" for i in range(len(observed))]
plt.figure(figsize=(8,5))
plt.bar(categories, observed, alpha=0.7, label="Observed", color="blue")
plt.bar(categories, expected, alpha=0.7, label="Expected", color="red", linestyle="dashed")
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Observed vs. Expected Frequencies")
plt.legend()
plt.grid(axis="y")
plt.show()
```
