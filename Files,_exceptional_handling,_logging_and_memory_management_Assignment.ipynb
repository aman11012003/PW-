### **2. Exception Handling in Python**
- Mechanism for managing runtime errors and preventing crashes.
- Involves `try`, `except`, `else`, and `finally` blocks.
- Simplifies debugging and ensures program stability.
- Handles specific or general exceptions.
- Improves user experience by providing meaningful error messages.

---

### **3. Purpose of the `finally` Block**
- Executes cleanup code regardless of exceptions.
- Ensures resources (e.g., files, connections) are properly closed.
- Provides a guaranteed code execution point.
- Often used with `try` and `except` blocks.
- Example: Closing a file after reading, even if an error occurs.

---

### **4. Logging in Python**
- Used to record events, errors, or messages during execution.
- Helps debug and monitor applications efficiently.
- Configurable through the `logging` module.
- Supports different log levels (e.g., DEBUG, INFO).
- Stores logs in files or outputs them to the console.

---

### **5. Significance of the `__del__` Method**
- Destructor method called when an object is garbage-collected.
- Performs cleanup tasks like closing files or releasing resources.
- Ensures memory is freed when an object is no longer in use.
- Rarely used due to Python's automatic memory management.
- Example: Deleting temporary files in a custom object.

````python
# 2 Open the file in read mode
# Ensure 'example.txt' exists in the same directory as the script
# or provide the full file path if it's located elsewhere.
# For example:
# with open('/path/to/your/file/example.txt', 'r') as file:
with open('output.txt', 'r') as file: # Change file name to 'output.txt'
    # Read and print each line in the file
    for line in file:
        print(line.strip())  # strip() removes leading/trailing whitespace
````

````python
#3 How would you handle a case where the file doesn't exist while trying to open it for reading?

try:
    # Attempt to open the file in read mode
    with open('example.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    # Handle the error if the file does not exist
    print("Error: The file does not exist.")
````

````python
#4 Write a Python script that reads from one file and writes its content to another file

try:
    # Open the source file in read mode and the target file in write mode
    with open('source.txt', 'r') as source_file, open('target.txt', 'w') as target_file:
        # Read the content of the source file
        content = source_file.read()

        # Write the content to the target file
        target_file.write(content)

    print("Content has been copied from source.txt to target.txt.")

except FileNotFoundError:
    # Handle the error if the file does not exist
    print("Error: The file 'source.txt' does not exist.")
````

````python
#5 How would you catch and handle division by zero error in Python
try:
    # Attempting division by zero
    result = 10 / 0
except ZeroDivisionError:
    # Handling the division by zero error
    print("Error: Cannot divide by zero.")
````

````python
#6 Log an error message for division by zero exception:


import logging
logging.basicConfig(filename='error.log', level=logging.ERROR)
try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error(f"Error occurred: {e}")
````
