
1. **What is NumPy, and why is it widely used in Python?**  
   NumPy is a powerful Python library for numerical computing. It provides support for multi-dimensional arrays, efficient element-wise operations, and tools for mathematical computations. Its performance is significantly faster than Python lists due to optimized C code and contiguous memory storage. NumPy is widely used in scientific computing, data analysis, and machine learning.

2. **How does broadcasting work in NumPy?**  
   Broadcasting in NumPy allows operations between arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one. This eliminates the need for explicit replication of data. Broadcasting simplifies code and improves efficiency by avoiding memory overhead.


3. **What is a Pandas DataFrame?**  
   A Pandas DataFrame is a two-dimensional, tabular data structure with labeled rows and columns. It is highly flexible and supports heterogeneous data types, making it ideal for data manipulation and analysis. DataFrames provide rich functionalities like filtering, grouping, and joining datasets.

4. **Explain the use of the `groupby()` method in Pandas.**  
   The `groupby()` method is used to split data into groups based on column values. After grouping, various aggregation functions like `sum()`, `mean()`, or custom functions can be applied to each group. This method is useful for summarizing and analyzing datasets efficiently.


5. **Why is Seaborn preferred for statistical visualizations?**  
   Seaborn simplifies the creation of statistically meaningful plots with its high-level API. It integrates well with Pandas and provides aesthetically pleasing plots like heatmaps, pair plots, and violin plots. Seaborn's default themes and color palettes make visualizations visually appealing with minimal customization.


6. **What are the differences between NumPy arrays and Python lists?**  
   NumPy arrays are faster and more memory-efficient due to their fixed data types and contiguous storage. Arrays support vectorized operations, enabling element-wise computations without explicit loops. In contrast, Python lists are more flexible but slower and lack built-in numerical operations.


7. **What is a heatmap, and when should it be used?**  
   A heatmap visualizes data in a matrix format, where values are represented using color intensity. It is ideal for identifying patterns, correlations, and trends in datasets, such as a correlation matrix or frequency table.


8. **What does the term “vectorized operation” mean in NumPy?**  
   Vectorized operations in NumPy perform element-wise computations on entire arrays without explicit Python loops. These operations are highly optimized and improve performance by leveraging low-level implementations.

9. **How does Matplotlib differ from Plotly?**  
   Matplotlib creates static, highly customizable visualizations suitable for scientific and publication-quality plots. Plotly focuses on interactive, web-based visualizations with features like zooming and dynamic tooltips. It is ideal for exploratory data analysis.


10. **What is the significance of hierarchical indexing in Pandas?**  
    Hierarchical indexing enables multi-level indexing, allowing complex datasets to be represented more intuitively. It simplifies slicing, grouping, and aggregating data along multiple dimensions, facilitating advanced data analysis tasks.

17. Explain the difference between apply() and map() in PandasA

   The apply() function applies a function to an entire DataFrame or Series along rows or columns. The map() function is used for element-wise transformations on a Pandas Series. While apply() is more flexible, map() is limited to Series and is generally faster for simple operations.

 Advanced features of NumPy include broadcasting, linear algebra functions, FFT (Fast Fourier Transform), random sampling, and memory mapping for large datasets. NumPy also supports custom data types and integration with other libraries like Pandas and SciPy.

20. What is the role of a pivot table in Pandas

  A pivot table summarizes and reorganizes data by grouping values based on specified indices and columns. It supports aggregation functions like sum, mean, or count, making it a powerful tool for summarizing large datasets.


21. **Why is NumPy’s array slicing faster than Python’s list slicing?**  

   NumPy’s array slicing is faster because it uses contiguous blocks of memory and operates at the C level, avoiding the overhead of Python’s dynamic type-checking. Slicing in NumPy creates views of the data without copying, whereas Python lists require new list creation and element-wise copying, making them slower.

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add a title
plt.title("Correlation Matrix Heatmap")

# Display the plot
plt.show()
```

```python
#Create a line plot with multiple lines using MatplotlibA
import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [3, 6, 9, 12, 15]

# Plot multiple lines
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-')
plt.plot(x, y2, label='Line 2', color='red', linestyle='--')
plt.plot(x, y3, label='Line 3', color='green', linestyle=':')

# Add title and labels
plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a thresholdA
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Aman', 'anand', 'rishu', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45]
}
df = pd.DataFrame(data)

# Set a threshold value
threshold = 30

# Filter rows where the 'Age' column is greater than the threshold
filtered_df = df[df['Age'] > threshold]

# Display the filtered DataFrame
print(filtered_df)

```

```python
# Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go

# Example data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [10, 11, 12, 13, 14]

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(size=10, color='blue', opacity=0.8)
)])

# Add title and labels
fig.update_layout(
    title="3D Scatter Plot",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
